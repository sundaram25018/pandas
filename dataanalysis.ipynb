```python
ax = sns.countplot(x = 'Order_Category',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x = 'Product_#',data = df)
sns.set(rc={'figure.figsize':(15,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_#'], as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=False).head(10)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data = sales_state, x = 'Product_#',y= 'Quantity')
```
